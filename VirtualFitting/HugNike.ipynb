```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time



Nike_Link = 'https://www.nike.com/t/sportswear-club-fleece-pullover-hoodie-Gw4Nwq/BV2654-751'

Hug_Link = 'https://huggingface.co/spaces/HumanAIGC/OutfitAnyone'

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import time
import os

# Set up Selenium WebDriver (Make sure to install the proper ChromeDriver for your browser)
driver = webdriver.Chrome()

# Navigate to the product page
driver.get(Nike_Link)

# Allow some time for the page to load
time.sleep(5)

# Locate the main image element using its unique data-testid
image_element = driver.find_element(By.CSS_SELECTOR, 'img[data-testid="HeroImg"]')

# Get the image URL from the src attribute
image_url = image_element.get_attribute('src')

# Download the image
image_data = requests.get(image_url).content

# Save the image locally
image_name = 'Images/nike_image.jpg'
with open(image_name, 'wb') as handler:
    handler.write(image_data)

print(f"Image downloaded successfully as {image_name}")

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time

# Set up Selenium WebDriver

Chrome_Options = Options()

Chrome_Options.add_argument("--headless")
Chrome_Options.add_argument("--disable-gpu")  # Disable GPU usage (not required, but recommended)
Chrome_Options.add_argument("--window-size=1920x1080")  # Set window size to avoid issues with element location
Chrome_Options.add_argument("--no-sandbox")  # Bypass OS security model (useful in certain environments)

driver = webdriver.Chrome(options=Chrome_Options)

# Open the Hugging Face page
driver.get(Hug_Link)

# Allow some time for the page to load
time.sleep(3)

# Debugging: Try locating the element by multiple methods

iframes = driver.find_elements(By.TAG_NAME, 'iframe')
print(f"Number of iframes on the page: {len(iframes)}")

for idx, iframe in enumerate(iframes):
    print(f"Iframe {idx}: {iframe.get_attribute('id') or iframe.get_attribute('name')}")

# If an iframe is found, switch to it
if iframes:
    driver.switch_to.frame(iframes[0])  # Switch to the first iframe (change index if necessary)
    print("Switched to the iframe.")

try:
    # Locate all input elements
    input_elements = driver.find_elements(By.TAG_NAME, 'input')
    print(f"Found {len(input_elements)} input elements.")

    # Loop through all input elements and print their types
    for idx, element in enumerate(input_elements):
        element_type = element.get_attribute('type')
        print(f"Input element {idx}: Type = {element_type}")
        
        if element_type == 'file':
            print("Found file input element.")
            file_path = '/Users/nadyashechtman/Desktop/VirtualFitting/Images/nike_image.jpg'  # Path to the file you want to upload
            element.send_keys(file_path)
            print('Upload initiated.')

            time.sleep(10)

            run_button = driver.find_element(By.ID, 'component-13')  # Adjust the ID if needed
            run_button.click()
            print('Run button clicked.')

            time.sleep(17)

            try:
                # Refined CSS selector to include the container and image
                image_container = driver.find_element(By.CSS_SELECTOR, 'div.image-container.svelte-1l6wqyv img.svelte-1pijsyv')
                image_url = image_container.get_attribute('src')
                print(f"Image URL: {image_url}")

                # Download the image using requests
                image_data = requests.get(image_url).content
                with open('FinalImage.webp', 'wb') as handler:
                    handler.write(image_data)
                print("Image downloaded successfully.")
            except Exception as e:
                print(f"Could not retrieve image: {e}")

            break
    else:
        print("No input element of type 'file' found.")

except Exception as e:
    print(f"Could not locate input elements: {e}")

driver.quit()
```
